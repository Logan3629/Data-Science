```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('airline cost new.csv')
x = df['num_pass']
y = df['cost']
df
```

```python
plt.scatter(x,y)
```

## Gradient Descent Algorithm

```python
m = 30
c = 1200

L = 0.0001         # L = Learning rate
epochs = 1000     # The count for reading the data
n = float(len(x))
for i in range(epochs):
    y_pred = m*x+c
    d_m = (-2/n)*sum(x*(y-y_pred))            # Cost function... Partial derivative is expressed
    d_c = (-2/n)*sum(y - y_pred)
    
    m = m - d_m * L
    c = c - d_c * L
    
    if (i % 100) == 0:
        plt.scatter(x,y)
        plt.plot(x, y_pred, 'r')
        plt.show()

print(m,c)
        
```
