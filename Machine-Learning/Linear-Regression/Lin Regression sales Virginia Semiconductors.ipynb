```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('virginia semiconductors.csv')
```

```python
df
```

```python
y = df['Size_of_Purchase']
x = df.drop('Size_of_Purchase', axis = 1)
```

```python
import statsmodels.api as sm
```

```python
xc = sm.add_constant(x)
```

```python
xc
```

```python
model = sm.OLS(y, xc).fit()
model.summary()
```

b_imports:
    There's 95% chance of getting |b_imports| >= 0.05 when H0 is true. Hence, we will accept H0 and conclude that 'imports' is not influencing size of purchase.

```python
y1 = df['Size_of_Purchase']
x1 = df.drop(['Size_of_Purchase', '%Imports', 'Distance'], axis = 1)
```

```python
xc1 = sm.add_constant(x1)
```

```python
xc1
```

```python
model1 = sm.OLS(y1, xc1).fit()
model1.summary()
```

Developed model for predicting size_of_purchase is Size_of_purchase = -8.5 +1.96*company_size*central_purchaser

Hence the model suggests that with a unit increase in company_size, we can expect additional sales of 1.196.

Moreover, if the buyer got central_purchase arrangement, we can get an additional of 126.9


R-Squared: 94.2 percent of variation in size_of_purchase is explained by company_size and central_purchaser (model).


F-Test:
  
  H0 b_company_size = b_central_purchaser = 0
  H1 b_company_size != b_central_purchaser != 0
  
2.09e - 0.9 is the probability of getting an Fvalue >= 114.7 when H0 is true.
Hence, there is no chance that we can get an Fvalue == 114.7 when H0 is true.

Hence, we will reject H0 to conclude that model as a whole is significant.
