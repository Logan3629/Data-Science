```python
x = pd.read_csv('x1_di.csv')
y = pd.read_csv('y1_di.csv')
```

```python
x.shape
```

```python
x = x.drop('Unnamed: 0',axis=1)
```

```python
y = y.drop('Unnamed: 0',axis=1)
```

```python
from sklearn.model_selection import  train_test_split

Xtrain,Xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=48,stratify=y)
```

```python
Xtrain.shape,Xtest.shape,ytrain.shape,ytest.shape
```

```python
from sklearn.preprocessing import  StandardScaler
st = StandardScaler()
Xtrain_t = st.fit_transform(Xtrain)
Xtrain_t=pd.DataFrame(Xtrain_t,columns=x.columns)
Xtrain_t.head()
```

```python
Xtest_t = st.transform(Xtest)
Xtest_t=pd.DataFrame(Xtest_t,columns=x.columns)
Xtest_t.head()
```

```python
from sklearn.metrics import classification_report,accuracy_score,roc_auc_score,roc_curve
```

# Logistic

```python
from sklearn.linear_model import LogisticRegression

Logistic = LogisticRegression()
```

# RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
RandomForest=RandomForestClassifier(**{'criterion': 'entropy',
 'max_depth': 9,
 'max_features': 7,
 'min_samples_leaf': 1,
 'min_samples_split': 14,
 'n_estimators': 143})
```

# AdaBoostClassifier + DT

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
wldt=DecisionTreeClassifier(max_depth=1,random_state=0)
AdaDt = AdaBoostClassifier(base_estimator=wldt, n_estimators=10,random_state=0)
```

# AdaBoostClassifier + LR

```python
from sklearn.ensemble import AdaBoostClassifier
wllr=LogisticRegression()
Adalr= AdaBoostClassifier(base_estimator=wllr, n_estimators=10,random_state=0)
```

# AdaBoostClassifier + RF

```python
wlrfc=RandomForestClassifier(**{'criterion': 'entropy',
 'max_depth': 9,
 'max_features': 7,
 'min_samples_leaf': 1,
 'min_samples_split': 14,
 'n_estimators': 143})
Adarf= AdaBoostClassifier(base_estimator=wlrfc, n_estimators=10,random_state=0)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=97)
```

# Naive base (Gaussian)

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
```

# lightgbm

```python
import lightgbm as lgb
lgbm=lgb.LGBMClassifier(**{'learning_rate': 0.18154188744813732,
 'max_depth': 9,
 'n_estimators': 95,
 'num_leaves': 20})
```

# GradientBoosting

```python
from sklearn.ensemble import GradientBoostingClassifier

gdBoost = GradientBoostingClassifier(**{'learning_rate': 0.37330572354222125,
 'max_depth': 2,
 'min_samples_leaf': 14,
 'min_samples_split': 14,
 'n_estimators': 164})
```

# XG boost

```python
from xgboost import XGBClassifier
xgBoost = XGBClassifier(**{'colsample_bytree': 0.44451235785409277,
 'gamma': 0.4785749927177465,
 'learning_rate': 0.0529230997548944,
 'max_depth': 9,
 'min_child_weight': 9,
 'n_estimators': 139})
```

# Voting (soft)

### (AdaBoost+RF) + LGBM + GradientBoost   

```python
from sklearn.ensemble import VotingClassifier
```

```python
stacked1 = VotingClassifier(estimators = [('AdaBoost+RF',Adarf),('LGBM', lgbm), ('GradientBoost', gdBoost)],voting='soft')
```

# (AdaBoost+RF) + LGBM

```python
stacked2 = VotingClassifier(estimators = [('AdaBoost+RF',Adarf),('LGBM', lgbm)],voting='soft')
```

# (AdaBoost+RF) +  GradientBoost

```python
stacked3 = VotingClassifier(estimators = [('AdaBoost+RF',Adarf), ('GradientBoost', gdBoost)],voting='soft')
```

# For all the model Bias~Varience analysis

###AUC

```python
from sklearn.model_selection import cross_val_score,KFold

kf = KFold(n_splits=10, shuffle=False, random_state=0)

model = [ Logistic  , RandomForest , AdaDt ,Adalr,Adarf,knn,gnb,lgbm , gdBoost , xgBoost ,stacked1,stacked2,stacked3]
model_name = ['logistic','RandomForest','AdaBoost+DT','AdaBoost+LR','AdaBoost+RF','KNN','Naive base','LGBM' , 'GradientBoost','XGboost',
              '(AdaBoost+RF)(LGBM)(GdB)','(AdaBoost+RF)(LGBM)','(AdaBoost+RF)(GdB)']


AUC_score_dic = {}
for i , j in zip(model,model_name):
  sc = cross_val_score(i , Xtrain_t , ytrain , scoring='roc_auc',cv=kf)
  AUC_score_dic[j]=sc

AUC_score_df = pd.DataFrame(AUC_score_dic)
AUC_score_df
```

```python
AUC_score_df.to_csv('AUC_score_df.csv')
```

```python
AUC_score_df.mean().sort_values(ascending=False)
```

```python
AUC_score_df1=AUC_score_df[list(AUC_score_df.median().sort_values().index)]
```

```python
plt.figure(figsize=(20,8))
AUC_score_df1.boxplot()
plt.ylabel('AUC Scores')
plt.xlabel('Models')
plt.xticks(rotation='vertical')
plt.show()
```

```python
bias =1-AUC_score_df.mean().sort_values(ascending=False)
bias
```

```python
bias[bias==bias.min()]
```

```python
bias_df = pd.DataFrame(bias).reset_index()
bias_df.columns = ['Models','Bias error']
```

```python
bias_df = bias_df.sort_values('Bias error',ascending=True)
bias_df
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x=bias_df['Models'],y = bias_df['Bias error']);
```

```python
variance = AUC_score_df.std()
variance.sort_values()
```

```python
plt.figure(figsize=(15,5))
variance.sort_values().plot(kind='bar')
plt.ylabel('Varience error')
plt.xlabel('Models')
plt.show()
```
